```python
def percorso_spesa(nodo, grafo, supermercati, nodi_percorsi, supermercato_trovato = False, ricerca_supermercato = False):
    # se non ci sono supermercati restituiamo errore
    if len(supermercati) < 1:
      return -1

    #print('nodo', nodo, ' nodi percorsi', nodi_percorsi, ' supermercato_trovato', supermercato_trovato)
    massima_lunghezza_percorso = float('inf') #inizializzo ad infinito per cercare il minimo
    
    # controllo se sono sul nodo destinazione
    if nodo == len(grafo):
        if supermercato_trovato == True:
          return 0 #destinazione trovata
        else: # caso in cui sono arrivato a destinazione ma non sono ancora passato da un supermercato
          ricerca_supermercato = True # cerco il supermercato partendo dal nodo destinazione

          # azzero la coda dei nodi percorsi aggiungendoci solo il nodo destinazione, altrimenti non potrei passare da nodi già percorsi
          nodi_percorsi.clear() 
          nodi_percorsi.append(nodo) 

    # serve per controllare se sono in un percorso valido, con via d'uscita
    percorso_valido = False 

    if nodo in supermercati :
        supermercato_trovato = True
        if ricerca_supermercato == True:
          # se stavo ricercando un supermercato dopo aver raggiunto destinazione, ho finito la ricerca
          return 0

    # usiamo nodo-1 perché i nodi sono numerati a partire da 1
    for i in range(len(grafo[nodo-1])):
        # controllo se il nodo 'i' è un vicino nel grafo del nodo 'nodo' e se non è ancora stato percorso
        if grafo[nodo-1][i]==1 and (i+1) not in nodi_percorsi :
            percorso_valido = True

            # aggiungo il nodo 'i' ai nodi percorsi
            nodi_percorsi_tmp = nodi_percorsi.copy()
            nodi_percorsi_tmp.append(i+1)

            lunghezza_percorso = percorso_spesa(i+1,grafo,supermercati,nodi_percorsi_tmp, supermercato_trovato, ricerca_supermercato)
            #print('nodo', nodo, ' vicino', i+1 , ' nodi_percorsi: ', nodi_percorsi, ' lunghezza_percorso=', lunghezza_percorso, ' best=', massima_lunghezza_percorso)

            # controllo se aggiornare la lunghezza massima del percorso
            if lunghezza_percorso < massima_lunghezza_percorso:
                massima_lunghezza_percorso = lunghezza_percorso
                #print('best nodo', nodo, ' aggiornato --> ', massima_lunghezza_percorso)

    if percorso_valido == False:
        # percorso non valido
        return float('inf') 


    if ricerca_supermercato == True and nodo == len(grafo):
      #print ('nodo terminale, lunghezza_percorso=', massima_lunghezza_percorso)
      # se sono sul nodo destinazione e ho dovuto ricercare il supermercato, allora i passi per raggiungere il supermercato
      # devono essere considerati 2 volte (andata e ritorno)
      return (massima_lunghezza_percorso+1)*2

    return massima_lunghezza_percorso +1
```
